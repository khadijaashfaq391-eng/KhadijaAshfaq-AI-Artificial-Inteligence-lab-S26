```python
data['Date_of_Journey'] = pd.to_datetime(data['Date_of_Journey'])

data['Journey_Day'] = data['Date_of_Journey'].dt.day
data['Journey_Month'] = data['Date_of_Journey'].dt.month

data.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
def convert_duration(x):
    h, m = 0, 0
    
    if 'h' in x:
        h = int(x.split('h')[0])
    if 'm' in x:
        m = int(x.split('m')[0].split()[-1])
        
    return h*60 + m

data['Duration'] = data['Duration'].apply(convert_duration)
```

```python
data['Dep_Hour'] = pd.to_datetime(data['Dep_Time']).dt.hour
data['Dep_Min'] = pd.to_datetime(data['Dep_Time']).dt.minute

data['Arrival_Hour'] = pd.to_datetime(data['Arrival_Time']).dt.hour
data['Arrival_Min'] = pd.to_datetime(data['Arrival_Time']).dt.minute

data.drop(['Dep_Time', 'Arrival_Time'], axis=1, inplace=True)
```

```python
data.drop(['Route'], axis=1, inplace=True)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib

# Dictionary to store all encoders
encoders = {}

# Columns to encode (categorical only)
categorical_cols = [
    'Airline',
    'Source',
    'Destination',
    'Total_Stops',
    'Additional_Info'
]

# Apply Label Encoding and store encoders
for col in categorical_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    encoders[col] = le

# Save encoders to file
joblib.dump(encoders, 'encoders.pkl')

print("Encoders saved successfully!")
```

```python
data.head()
```

```python
data.info()
```

```python
data['Price'] = data['Price'].fillna(data['Price'].mean())
```

```python
def fillnaMean(cols):
    for i in cols:
        data[i] = data[i].fillna(data[i].mean())

columns = ['Price']

fillnaMean(columns)
```

```python
data.info()
```

```python
data['Price'] = data['Price'].replace(',', '', regex=True).astype(float)
```

```python
import pandas as pd
import numpy as np
import pickle
import joblib

# Load Saved Model & Encoders
model = joblib.load(open('model.pkl', 'rb'))
encoders = joblib.load('encoders.pkl')

# Function for Encoding
def encode_cols(le, value):
    if value in le.classes_:
        return le.transform([value])[0]
    else:
        return -1

print("----- Flight Fare Prediction -----")

# User Inputs
user_input = {
    
    "Airline": input("Airline: "),
    "Source": input("Source: "),
    "Destination": input("Destination: "),
    
    "Duration": int(input("Duration (in minutes): ")),
    
    "Total_Stops": input("Total Stops: "),
    
    "Additional_Info": input("Additional Info: "),
    
    "Journey_Day": int(input("Journey Day: ")),
    "Journey_Month": int(input("Journey Month: ")),
    
    "Dep_Hour": int(input("Departure Hour (0-23): ")),
    "Dep_Min": int(input("Departure Minute (0-59): ")),
    
    "Arrival_Hour": int(input("Arrival Hour (0-23): ")),
    "Arrival_Min": int(input("Arrival Minute (0-59): "))
}

# Encoding Categorical Columns
categorical_cols = [
    'Airline',
    'Source',
    'Destination',
    'Total_Stops',
    'Additional_Info'
]

for col in categorical_cols:
    user_input[col] = encode_cols(encoders[col], user_input[col])

# Features Order (same as training)
features = [
    'Airline',
    'Source',
    'Destination',
    'Duration',
    'Total_Stops',
    'Additional_Info',
    'Journey_Day',
    'Journey_Month',
    'Dep_Hour',
    'Dep_Min',
    'Arrival_Hour',
    'Arrival_Min'
]

# Convert Input into Array
X = np.array([[user_input[col] for col in features]])

# Predict Flight Fare
prediction = model.predict(X)

print("\nPredicted Flight Fare:", prediction[0])
```
