1. DFS with Stack & Node

```python
def dfs_stack(graph, root):

    visited = set()
    stack = [root]

    while stack:
        vertex = stack.pop()

        if vertex not in visited:
            print(str(vertex) + " ", end="")
            visited.add(vertex)

            for neighbour in graph[vertex]:
                if neighbour not in visited:
                    stack.append(neighbour)


if __name__ == '__main__':

    graph = {0: [1, 2],
             1: [2], 
             2: [3], 
             3: [1, 2]}

    print("Following is Depth First Traversal using Stack: ")
    dfs_stack(graph, 0)
```

2. Research about "Inorder, Preorder, Postorder" and implement in DFS 

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

# Preorder

def preorder(root):

    if root:
        print(str(root.data) + " ", end="")
        preorder(root.left)
        preorder(root.right)

# Inorder 

def inorder(root):

    if root:
        inorder(root.left)
        print(str(root.data) + " ", end="")
        inorder(root.right)

# Postorder

def postorder(root):

    if root:
        postorder(root.left)
        postorder(root.right)
        print(str(root.data) + " ", end="")

if __name__ == '__main__':

    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)

    print("Preorder Traversal:")
    preorder(root)

    print("Inorder Traversal:")
    inorder(root)

    print("Postorder Traversal:")
    postorder(root)                        
```
