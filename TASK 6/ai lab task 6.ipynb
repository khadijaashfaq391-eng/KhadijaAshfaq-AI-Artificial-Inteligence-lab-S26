1. BFS without Queue & without Node

```python
# BFS without Queue & without Node

graph = {

   'A':['B', 'C'],
   'B':['D', 'E'],
   'C':['F'],
   'D':['G','H'],
   'E':[],
   'F':['I','K'],
   'G':[],
   'H':['L'],
   'I':[],
   'K':['M'],
   'L':[],
   'M':[]

}

def bfs(start):

    visited = []
    queue = []  

    visited.append(start)
    queue.append(start)

    while queue:

        vertex = queue.pop(0)
        print(vertex)

        for child in graph[vertex]:
            if child not in visited:
                visited.append(child)
                queue.append(child)


start = 'A'
bfs(start)
```

2. BFS with Queue & Node 

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.children = []

# BFS with Queue & Node

def bfs(root):

    visited = []
    queue = []

    visited.append(root)
    queue.append(root)

    while queue:

        node = queue.pop(0)
        print(node.data)

        for child in node.children:
            if child not in visited:
                visited.append(child)
                queue.append(child)

A = Node('A')
B = Node('B')
C = Node('C')
D = Node('D')
E = Node('E')
F = Node('F')

A.children = [B, C]
B.children = [D, E]
C.children = [F]

bfs(A)        
```
