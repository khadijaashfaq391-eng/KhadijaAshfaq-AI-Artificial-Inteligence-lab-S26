1. Code for LUHN Algorithm

```python
print("LUHN Algorithm Checker")

card_number = input("Enter card number: ")
check_digit = int(card_number[-1])
digits = card_number[:-1]

digits = digits[::-1]
total = 0

for i in range(len(digits)):
    num = int(digits[i])
    
    if i % 2 == 0:
        num = num * 2
        if num > 9:
            num = num - 9
    
    total += num

total += check_digit

if total % 10 == 0:
    print("Valid Card Number")
else:
    print("Invalid Card Number")
```

2. Write a python program to remove punctuations from the given string? 

```python
print("Remove Punctuations")

text = input("Enter a sentence: ")

punctuations = ".,!?;:'()"

new_text = ""

for ch in text:
    if ch not in punctuations:
        new_text += ch

print("Without punctuation:", new_text)
```

3. Write a python program to sort the sentence in alphabetical order? 

```python
print("Sort Sentence Alphabetically")

sentence = input("Enter a sentence: ")

words = sentence.split()

words.sort()

print("Sorted sentence:")
for word in words:
    print(word)
```
